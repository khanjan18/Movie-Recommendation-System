```python
def get_director(x):
    if '(director/screenplay);' in x:
        return x.split('(director/screenplay);')[0]
    elif '(director/acreenplay);' in x:
        return x.split('(director/acreenplay);')[0]
    elif '(directors/screenplay);' in x:
        return x.split('(directors/screenplay);')[0]
    elif '(director/screenplay)' in x:
        return x.split('(director/screenplay)')[0]
    elif '(director/sceenplay);' in x:
        return x.split('(director/sceenplay);')[0]
    elif '(director/screenwriter)' in x:
        return x.split('(director/screenwriter)')[0]
    elif '(director);' in x:
        return x.split('(director);')[0]
    elif '(director)' in x:
        return x.split('(director)')[0]
    else:
        return x.split('(directors);')[0]
```

```python
movies_2021['director_name'] = movies_2021['Cast and crew'].map(lambda x: get_director(str(x)))
```

```python
movies_2021['director_name'] = movies_2021['director_name'].map(lambda x: get_director(x))
```

```python
def get_actor_1_name(x):
    if len (x.split('y); ')[-1].split(', ')) < 1:
        return np.NaN
    elif len (x.split('y); ')[-1].split(', ')) >= 1:
        return x.split('y); ')[-1].split(', ')[0]
    elif len (x.split('s); ')[-1].split(', ')) < 1:
        return np.NaN
    elif len (x.split('s); ')[-1].split(', ')) >= 1:
        return x.split('s); ')[-1].split(', ')[0]
    elif len (x.split('r); ')[-1].split(', ')) < 1:
        return np.NaN
    else:
        return x.split('r); ')[-1].split(', ')[0]
```

```python
movies_2021['actor_1_name'] = movies_2021['Cast and crew'].map(lambda x: get_actor_1_name(str(x)))
```

```python
def get_actor_2_name(x):
    if len (x.split('y); ')[-1].split(', ')) < 2:
        return np.NaN
    elif len (x.split('y); ')[-1].split(', ')) >= 2:
        return x.split('y); ')[-1].split(', ')[1]
    elif len (x.split('s); ')[-1].split(', ')) < 2:
        return np.NaN
    elif len (x.split('s); ')[-1].split(', ')) >= 2:
        return x.split('s); ')[-1].split(', ')[1]
    elif len (x.split('r); ')[-1].split(', ')) < 2:
        return np.NaN
    else:
        return x.split('r); ')[-1].split(', ')[1]
```

```python
movies_2021['actor_2_name'] = movies_2021['Cast and crew'].map(lambda x: get_actor_2_name(str(x)))
```

```python
def get_actor_3_name(x):
    if len (x.split('y); ')[-1].split(', ')) < 3:
        return np.NaN
    elif len (x.split('y); ')[-1].split(', ')) >= 3:
        return x.split('y); ')[-1].split(', ')[2]
    elif len (x.split('s); ')[-1].split(', ')) < 3:
        return np.NaN
    elif len (x.split('s); ')[-1].split(', ')) >= 3:
        return x.split('s); ')[-1].split(', ')[2]
    elif len (x.split('r); ')[-1].split(', ')) < 3:
        return np.NaN
    else:
        return x.split('r); ')[-1].split(', ')[2]
```

```python
movies_2021['actor_3_name'] = movies_2021['Cast and crew'].map(lambda x: get_actor_3_name(str(x)))
```

```python
from tmdbv3api import TMDb
import json
import requests
tmdb = TMDb()
tmdb.api_key = '83c52c49ddee44475891ab5d3a3e5a3d'
from tmdbv3api import Movie
tmdb_movie = Movie() 
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    if not result:
        return np.NaN
    else:
        movie_id = result[0].id
        response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
        data_json = response.json()
        if data_json['genres']:
            genre_str = " " 
            for i in range(0,len(data_json['genres'])):
                genres.append(data_json['genres'][i]['name'])
            return genre_str.join(genres)
        else:
            return np.NaN
```

```python
movies_2021['genres'] = movies_2021['Title'].map(lambda x: get_genre(str(x)))
```
