```python
import numpy as np
import pandas as pd
```

```python
final_data = pd.read_csv('../datasets/final_data.csv')
```

```python
final_data.head()
```

```python
#CountVectorizer counts text(TAGS) converted into vectors.
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 6000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(final_data['tags']).toarray()
```

```python
vectors
```

```python
vectors.shape
```

```python
cv.get_feature_names()
```

```python
#Stem words
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stemmer(x):
    stem_word = []
    string = " "
    
    for i in x.split():
        stem_word.append(ps.stem(i))
        
    return string.join(stem_word)
```

```python
final_data['tags'] = final_data['tags'].map(lambda x: stemmer(x))
```

```python
vectors = cv.fit_transform(final_data['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
#Cosine Similarity measures similarity between two vectors.
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
def recommend(movie):
    movie_index = final_data[final_data['movie_title'] == movie].index[0]
    distances = similarity[movie_index]
    recommended_movies_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x: x[1])[1:9]
    
    for i in recommended_movies_list:
        print(final_data.iloc[i[0]].movie_title)
```

```python
recommend('spider-man: no way home')
```
