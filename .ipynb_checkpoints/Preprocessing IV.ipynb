```python
import numpy as np
import pandas as pd
```

## Extracting 2020 Movies

```python
link = "https://en.wikipedia.org/wiki/List_of_American_films_of_2020"
df1 = pd.read_html(link, header = 0)[2]
df2 = pd.read_html(link, header = 0)[3]
df3 = pd.read_html(link, header = 0)[4]
df4 = pd.read_html(link, header = 0)[5]
```

```python
movies_2020 = df1.append(df2.append(df3.append(df4,ignore_index = True),ignore_index = True),ignore_index =True)
```

```python
movies_2020
```

```python
def get_director(x):
    if ' (director)' in x:
        return x.split(" (director)")[0]
    elif ' (directors)' in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
```

```python
def get_actor_1_name(x):
    return (x.split("y); ")[-1].split(", ")[0])
```

```python
def get_actor_2_name(x):
    if len((x.split("y); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("y); ")[-1]).split(", ")[1])
```

```python
def get_actor_3_name(x):
    if len(x.split(" (screenplay);")[-1].split(", ")) < 3:
        return np.NaN
    else:
        return (x.split(" (screenplay)")[-1].split(", ")[2])
```

```python
movies_2020['director_name'] = movies_2020['Cast and crew'].map(lambda x: get_director(x))
```

```python
movies_2020['actor_1_name'] = movies_2020['Cast and crew'].map(lambda x: get_actor_1_name(x))
```

```python
movies_2020['actor_2_name'] = movies_2020['Cast and crew'].map(lambda x: get_actor_2_name(x))
```

```python
movies_2020['actor_3_name'] = movies_2020['Cast and crew'].map(lambda x: get_actor_3_name(x))
```

```python
from tmdbv3api import TMDb
import json
import requests
tmdb = TMDb()
tmdb.api_key = '83c52c49ddee44475891ab5d3a3e5a3d'
from tmdbv3api import Movie
tmdb_movie = Movie() 
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    if not result:
        return np.NaN
    else:
        movie_id = result[0].id
        response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
        data_json = response.json()
        if data_json['genres']:
            genre_str = " " 
            for i in range(0,len(data_json['genres'])):
                genres.append(data_json['genres'][i]['name'])
            return genre_str.join(genres)
        else:
            return np.NaN
```

```python
movies_2020['genres'] = movies_2020['Title'].map(lambda x: get_genre(str(x)))
```

```python
movies_2020.rename(columns= {'Title': 'movie_title'}, inplace = True)
```

```python
movies_2020['movie_title'] = movies_2020['movie_title'].str.lower()
```

```python
movies_2020 = movies_2020.loc[:,['genres','movie_title','director_name','actor_1_name','actor_2_name','actor_3_name']]
```

```python
movies_2020['tags'] = movies_2020['genres']+" "+movies_2020['movie_title']+" "+movies_2020['director_name']+" "+movies_2020['actor_1_name']+" "+movies_2020['actor_2_name']+" "+movies_2020['actor_3_name']
```
