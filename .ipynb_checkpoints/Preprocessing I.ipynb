```python
# Importing Libraries
import pandas as pd
import numpy as np
```

```python
#dataset
metadata = pd.read_csv('datasets/movie_metadata.csv')
```

```python
pd.set_option('display.max_columns', None)
metadata.head()
```

```python
metadata['title_year'].value_counts().sort_index() #Number of movies released in particular year 
```

```python
#Columns which are important 
metadata = metadata[['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
metadata.head()
```

```python
metadata['genres'] = metadata['genres'].str.replace('|',' ')
```

```python
metadata.head()
```

```python
metadata['movie_title'][0]
```

```python
metadata['movie_title'] = metadata['movie_title'].apply(lambda x: x[:-1])
```

```python
metadata['movie_title'][0]
```

```python
#Getting movies' details Using TMDB API. 
from tmdbv3api import TMDb
import json
import requests
tmdb = TMDb()
tmdb.api_key = '83c52c49ddee44475891ab5d3a3e5a3d'
from tmdbv3api import Movie
tmdb_movie = Movie() 
def get_overview(x):
    overviews = []
    result = tmdb_movie.search(x)
    movie_id = result[0].id
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
    data_json = response.json()
    overview = " "
    overviews.append(data_json['overview'])
    return overview.join(overviews)

metadata['overview'] = metadata['movie_title'].map(lambda x: get_overview(str(x)))
```

```python
#Replace Null values with string
metadata['actor_1_name'] = metadata['actor_1_name'].replace(np.NaN, 'unknown')
metadata['actor_2_name'] = metadata['actor_2_name'].replace(np.NaN, 'unknown')
metadata['actor_3_name'] = metadata['actor_3_name'].replace(np.NaN, 'unknown')
metadata['director_name'] = metadata['director_name'].replace(np.NaN, 'unknown')
```

```python
#Saving data in CSV format
metadata.to_csv('till_2016_data.csv', index = False)
```
