```python
import numpy as np
import pandas as pd
```

```python
#Dataset for 2017 movies.
credits = pd.read_csv('datasets/credits.csv')
credits
```

```python
#Dataset for 2017 movies.
meta = pd.read_csv('datasets/movies_metadata.csv')
```

```python
#Converting into datetime format to get year
meta['release_date'] = pd.to_datetime(meta['release_date'], errors='coerce')
```

```python
meta['year'] = meta['release_date'].dt.year
```

```python
meta['year'].value_counts().sort_index()
```

```python
# Getting only 2017 movies as we already have movies up to the year 2016 in preprocessing 1 file. 
# We don't have enough data for the movies from 2018, 2019 and 2020. 
# We'll deal with it in the upcoming preprocessing files
new_meta = meta.loc[meta.year == 2017,['genres','id','title','year']]
new_meta
```

```python
new_meta['id'] = new_meta['id'].astype(int)
data = pd.merge(new_meta, credits, on='id')
pd.set_option('display.max_colwidth', 75)
data
```

```python
import ast
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x))
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
```

```python
data.head()
```

```python
def genres_list(x):
    genres = []
    string = " "
    for i in x:        
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            genres.append(scifi)
        else:
            genres.append(i.get('name'))
    if genres == []:
        return np.NaN
    else:
        return (string.join(genres)) 
```

```python
data['genres'] = data['genres'].map(lambda x: genres_list(x))
```

```python
data
```

```python
data['cast'][0]
```

```python
def get_actor_name_1(x):
    actor_name = []
    for i in x:
        actor_name.append(i.get('name'))
    if actor_name == []:
        return np.NaN
    else:
        return actor_name[0]
```

```python
data['actor_name_1'] = data['cast'].map(lambda x: get_actor_name_1(x))
```

```python
data.head()
```

```python
def get_actor_name_2(x):
    actor_name = []
    for i in x:
        actor_name.append(i.get('name'))
    if actor_name == [] or len(actor_name) <= 1:
        return np.NaN
    else:
        return actor_name[1]
```

```python
data['actor_name_2'] = data['cast'].map(lambda x: get_actor_name_2(x))
```

```python
data.head()
```

```python
def get_actor_name_3(x):
    actor_name = []
    for i in x:
        actor_name.append(i.get('name'))
    if actor_name == [] or len(actor_name) <= 2:
        return np.NaN
    else:
        return actor_name[2]
```

```python
data['actor_name_3'] = data['cast'].map(lambda x: get_actor_name_3(x))
```

```python
data
```

```python
def get_director(x):
    director_name = []
    string = " "
    for i in x:
        if i.get('job') == 'Director':
            director_name.append(i.get('name'))
    if director_name == []:
        return np.NaN
    else:
        return (string.join(director_name))
```

```python
data['director_name'] = data['crew'].map(lambda x: get_director(x))
```

```python
data.head()
```

```python
data = data.loc[:, ['genres','title','actor_name_1','actor_name_2','actor_name_3','director_name']]
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data.dropna(how = 'any', inplace = True)
```

```python
data.isna().sum()
```

```python
data['title'] = data['title'].str.lower()
```

```python
data['tags'] = data['genres'] + " "+ data['title']+ " "+ data['actor_name_1'] + " "+data['actor_name_2'] + " " + data['actor_name_3'] + data['director_name']
```
